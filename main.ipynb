# COVID-19 in 2020

Few years ago, our lives were so different compared to what they are now, workers principally went to work to the physical place of their company, students were is school learning using books while looking at the board, and in free time they were enjoying while playing with their friends, now workers do their job at home with a computer while students are connected to videoconference programms to listen their teachers. The lifestyle of the entire planet changed dramatically when COVID-19 begun its spreading across the world, threating the health of every country who got it.

COVID is the disease caused by SARS-CoV-2, the coronavirus that emerged in December 2019. It spreads through the air, when a infected person talks, breathes or expels matter from the mouth or nose, and it is particularly infectious in areas where are many people and the hygiene is poor. For that reason, the use of mask has become fundamental in the fight again the desease.

Despite that the entire world use quarantine and wearing mask as the main techniques against COVID, it is well known that every country has its own approach to minimize its progress, and the reason of this is because not every country lives the same situation, in economical and social matter.

The purpose of this document is to observe the progress of COVID around the globe, paying attention to the biggest contries in terms of population and focusing in its spread through Southamerica, comparing the situation between countries.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv("country_wise_latest.csv",header=0)
dataset.sort_values(by=['Confirmed'],ascending=False,inplace=True)
day_wise = pd.read_csv("day_wise.csv",header=0)
```

### COVID around the world

To begin with, is necessary to see how the COVID-19 propagated through the globe, in matter of people confirmed with the desease, the number of people recovered and and the amount of people who died by the desease.

```python
fg,ax = plt.subplots(1,2,figsize=(16,7))
dates = matplotlib.dates.date2num(day_wise['Date'])
ax[1].plot_date(dates,day_wise['Confirmed'],label='Confirmed',linestyle='solid',marker='')
ax[1].plot_date(dates,day_wise['Recovered'],label='Recovered',linestyle='solid',marker='')
ax[1].plot_date(dates,day_wise['Deaths'],label='Deaths',linestyle='solid',marker='')
ax[1].legend(loc="upper left")
ax[1].set_title('COVID cases around the world in February-August 2020')

ax[0].plot_date(dates,day_wise['No. of countries'],label='No. of countries',linestyle='solid',marker='')
ax[0].legend(loc="upper left")
ax[0].set_title('Countries affected by COVID in February-August 2020')
```

Is noticeable how in February the spread of COVID slow down until March, and that may be because the majority of countries were aware recently of the potential threat COVID is and they took the most common measure for this kind of cases, quarantine. But it was until May when every country on the planet were infected with the virus.

## COVID-19 in Chile and Southamerica

Is fundamental to observe what was the situation of Chile in terms of time, how the desease spread in the country and how the situation evolved, and that is the purpose of the next graphic:

```python
dataset_grouped = pd.read_csv("full_grouped.csv",header=0)

fg2, ax2 = plt.subplots(1,figsize=(15,7))

data_chile = dataset_grouped.loc[dataset_grouped['Country/Region']=='Chile']
dates = matplotlib.dates.date2num(data_chile['Date'])
ax2.plot_date(dates,data_chile['Confirmed'],label='Confirmed',linestyle='solid',marker='')
ax2.plot_date(dates,data_chile['Recovered'],label='Recovered',linestyle='solid',marker='')
ax2.plot_date(dates,data_chile['Deaths'],label='Deaths',linestyle='solid',marker='')
ax2.set_title('COVID situation in Chile 2020')
ax2.legend(loc="upper left")
fg2.savefig('COVID_Chile.jpg')

```

A peculiarity that can be found in the previous image, is that recovered line is so close to the confirmed line, it could mean that a high percentage of people who got COVID-19 recovered from it. In comparison, the following graphic shows how was te situation in US in 2020, using the same metrics to compare: confirmed, recovered and deaths.

```python
data_chile = dataset_grouped.loc[dataset_grouped['Country/Region']=='Chile']
dates = matplotlib.dates.date2num(data_chile['Date'])
fg3, ax3 = plt.subplots(1,2,figsize=(18,6),sharey=True)
ax3[0].plot_date(dates,data_chile['Confirmed'],label='Confirmed',linestyle='solid',marker='')
ax3[0].plot_date(dates,data_chile['Recovered'],label='Recovered',linestyle='solid',marker='')
ax3[0].plot_date(dates,data_chile['Deaths'],label='Deaths',linestyle='solid',marker='')
ax3[0].set_title('COVID situation in Chile 2020')
ax3[0].legend(loc="upper left")

data_us = dataset_grouped.loc[dataset_grouped['Country/Region']=='US']

ax3[1].plot_date(dates,data_us['Confirmed'],label='Confirmed',linestyle='solid',marker='')
ax3[1].plot_date(dates,data_us['Recovered'],label='Recovered',linestyle='solid',marker='')
ax3[1].plot_date(dates,data_us['Deaths'],label='Deaths',linestyle='solid',marker='')
ax3[1].set_title('COVID situation in US 2020')
ax3[1].legend(loc="upper left")
```

Besides US have more people who got sick of COVID-19 (because it has a higher population than Chile), is noticeable that the level of people who recoreved from COVID-19 in comparison with amount of people confirmed was much lower than in Chile, but is quite hard to see it. The objective of the following picture is to allow us to see the diference between both countries.

```python
data_chile = dataset_grouped.loc[dataset_grouped['Country/Region']=='Chile']
data_us = dataset_grouped.loc[dataset_grouped['Country/Region']=='US']

fg4,ax4 = plt.subplots(figsize=(12,6))
ax4.plot_date(dates,(data_chile['Recovered']/data_chile['Confirmed']),label='Recovered/Confirmed in Chile',linestyle='solid',marker='')
ax4.plot_date(dates,(data_us['Recovered']/data_us['Confirmed']),label='Recovered/Confirmed in US',linestyle='solid',marker='')

ax4.legend(loc="upper left")
```

While US started strong in terms of recovery, reaching a peak over 0.4 in Recovered/Confirmed ratio, it drops down at the beginning of March, while Chile started going up at mid-March. With that information, it can be said that despite US have more people recovered, in proportion Chile have done a best work against COVID-19 over time.

```python
def country_plot(x,y,ax,country,df,population=0):
    df_target = df.loc[df['Country/Region']==country]
    dates = matplotlib.dates.date2num(df_target['Date'])
    if population:
        df_target['Confirmed'] = df_target['Confirmed']/population
        df_target['Recovered'] = df_target['Recovered']/population
        df_target['Deaths'] = df_target['Deaths']/population
        ax[x][y].plot_date(dates,df_target['Confirmed'],label='Confirmed',linestyle='solid',marker='')
        ax[x][y].plot_date(dates,df_target['Recovered'],label='Recovered',linestyle='solid',marker='')
        ax[x][y].plot_date(dates,df_target['Deaths'],label='Deaths',linestyle='solid',marker='')
        ax[x][y].set_title('COVID situation in '+ country +' 2020')
        ax[x][y].legend(loc="upper left")
    else:
        ax[x][y].plot_date(dates,df_target['Confirmed'],label='Confirmed',linestyle='solid',marker='')
        ax[x][y].plot_date(dates,df_target['Recovered'],label='Recovered',linestyle='solid',marker='')
        ax[x][y].plot_date(dates,df_target['Deaths'],label='Deaths',linestyle='solid',marker='')
        ax[x][y].set_title('COVID situation in '+ country +' 2020')
        ax[x][y].legend(loc="upper left")
```

Now focusing in the neighborhood of Chile, Southamerica have its own situation with the Coronavirus, the environment of every country is different, in matter of politics, economics and demographics. The image below shows the number of people per country affected with the desease.

```python
fg4, ax4 = plt.subplots(3,3,figsize=(15,15),sharey=True)
data_sample = dataset_grouped.loc[dataset_grouped['Country/Region']=='Chile']

country_plot(0,0,ax4,'Chile',dataset_grouped)
country_plot(0,1,ax4,'Argentina',dataset_grouped)
country_plot(0,2,ax4,'Peru',dataset_grouped)
country_plot(1,0,ax4,'Bolivia',dataset_grouped)
country_plot(1,1,ax4,'Paraguay',dataset_grouped)
country_plot(1,2,ax4,'Uruguay',dataset_grouped)
country_plot(2,0,ax4,'Ecuador',dataset_grouped)
country_plot(2,1,ax4,'Venezuela',dataset_grouped)
country_plot(2,2,ax4,'Colombia',dataset_grouped)

#Se puede ver los afectados en cantidad de personas, pero no a nivel pas.
```

It seems Chile and Argentina leads in Southamerica in regard to the amount of people affected by the desease, but it is time to see how every country is doing with respect to the total population.

```python
fg5, ax5 = plt.subplots(3,3,figsize=(18,18),sharey=True)
data_sample = dataset_grouped.loc[dataset_grouped['Country/Region']=='Chile']

country_plot(0,0,ax5,'Chile',dataset_grouped,19116201)
country_plot(0,1,ax5,'Argentina',dataset_grouped,45195777)
country_plot(0,2,ax5,'Peru',dataset_grouped,33149016)
country_plot(1,0,ax5,'Bolivia',dataset_grouped,11633371)
country_plot(1,1,ax5,'Paraguay',dataset_grouped,7353672)
country_plot(1,2,ax5,'Uruguay',dataset_grouped,3518552)
country_plot(2,0,ax5,'Ecuador',dataset_grouped,17803339)
country_plot(2,1,ax5,'Venezuela',dataset_grouped,28515829)
country_plot(2,2,ax5,'Colombia',dataset_grouped,51049498)
```

One reason why there is a breach of diference between results may be because every country took a different approach about how to treat the COVID-19, like it was said earlier, not everyone have the same conditions and resources to fight the progress of the desease. Also the population density can affect this too, because if it is too high, it means there is a high level of concentration of people in a specific geographic locale, leading to a high chance to spread the COVID-19 faster. One last cause could be that not every person who got the disease went to the hospital and got counted.
